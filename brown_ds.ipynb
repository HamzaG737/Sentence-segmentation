```python
#nltk.download('punkt')
import numpy as np
from collections import defaultdict
import random
```

```python
from nltk.corpus import brown
brown_sents = brown.sents()
len(brown_sents)
```

```python
## longueur moyenne des phrases : 
mean_=0
for sent in brown_sents : 
    mean_ += len(sent)
mean_/len(brown_sents)
```

## Change half samples by removing dot 

```python
change_sents = []
for sent in brown_sents :
    if sent[-1]=='.' : 
        k = np.random.binomial(size=1, n=1, p= 0.5)
        if k == 0 : 
            change_sents.append(sent)
        else : 
            if len(sent[:-1])==0 : 
                continue
            else : 
                change_sents.append(sent[:-1])
    else :
        change_sents.append(sent)
```

```python
final_chars = defaultdict(int)
for sent in change_sents :
    try : 
        final_chars[sent[-1]]+=1
    except :
        print(len(sent))
dict(final_chars)  
```

## create labels 
1 : for the end of a sentence and 0 elsewhere. 

```python
tab_labels = []
for sent in change_sents :
    labels = [0] * len(sent)
    labels[-1] = 1 
    tab_labels.append(labels)
```

## Constitute dataset 

We regroup each K sentences in one sample.

```python
K = random.randrange(1,6)
len_dataset = len(change_sents) // K
grouped_sent , grouped_labels = []  , [] 
p=0
for i, (sent,label) in enumerate(zip(change_sents,tab_labels),1) :
    try : 
        grouped_sent[p]+=sent
        grouped_labels[p]+=label
    except : 
        grouped_sent.append(sent)
        grouped_labels.append(label)
    if i%K ==0 :
        p+=1
        K = random.randrange(1,10)
        
        
```
